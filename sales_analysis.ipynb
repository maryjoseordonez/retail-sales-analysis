```python
import matplotlib.pyplot as plt

df.groupby("Region")["Sales"].sum().sort_values().plot(kind="barh")

plt.title("Sales by Region")
plt.xlabel("Total Sales")
plt.ylabel("Region")

plt.show()
```

```python
df.groupby("Product.Name")["Profit"].sum().sort_values(ascending=False).head(10)
```

```python
df["Order.Date"] = pd.to_datetime(df["Order.Date"])
```

```python
monthly_sales = df.groupby(df["Order.Date"].dt.to_period("M"))["Sales"].sum()
monthly_sales
```

```python
monthly_sales.plot(figsize=(12,6), title="Monthly Sales Trend")
```

```python
df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)
```

```python
df.groupby("Segment")["Sales"].sum().plot(kind="bar", title="Sales by Customer Segment")
```

```python
customer_sales = df.groupby("Customer.Name")["Sales"].sum().sort_values(ascending=False)

customer_sales.head(10)
```

```python
customer_frequency = df.groupby("Customer.Name")["Order.ID"].count().sort_values(ascending=False)

customer_frequency.head(10)
```

```python
customer_avg_value = df.groupby("Customer.Name")["Sales"].mean().sort_values(ascending=False)

customer_avg_value.head(10)
```

```python
top_customers = df.groupby("Customer.Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_customers.plot(kind="bar", title="Top 10 Customers by Revenue")
```
